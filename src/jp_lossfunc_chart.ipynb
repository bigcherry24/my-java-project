```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 예제 데이터 생성
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# 손실 함수 (평균제곱오차)
def mse_loss(theta0, theta1, X, y):
    m = len(y)
    predictions = theta0 + theta1 * X
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

# theta0, theta1 값 생성
theta0_vals = np.linspace(-5, 5, 100)
theta1_vals = np.linspace(-5, 5, 100)
theta0_vals, theta1_vals = np.meshgrid(theta0_vals, theta1_vals)
loss_vals = np.zeros_like(theta0_vals)

# 손실 함수 값 계산
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        loss_vals[i, j] = mse_loss(theta0_vals[i, j], theta1_vals[i, j], X, y)

# 3D 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(theta0_vals, theta1_vals, loss_vals, cmap='viridis')

# 그래프 레이블 설정
ax.set_xlabel('Theta0')
ax.set_ylabel('Theta1')
ax.set_zlabel('Loss (MSE)')

plt.show()
```
